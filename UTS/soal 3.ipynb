```python
def list_employees_by_domain():
    company_domain_map = {company['name']: company['domain'] for company in company_detail_list}
    
    employees_by_domain = {}
    
    sorted_employee_list = sorted(employee_detail_list, key=lambda x: x['name'])
    
    for employee in sorted_employee_list:
        company = employee['company']
        domain = company_domain_map.get(company, 'Unknown')
        full_name = employee['full_name']
        
        if domain not in employees_by_domain:
            employees_by_domain[domain] = []
        
        employees_by_domain[domain].append(full_name)
    
    return employees_by_domain
```

```python
output = list_employees_by_domain()
```
